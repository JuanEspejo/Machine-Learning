###  1. Leer el archivo .cdf

```python
#Librerias
import netCDF4 as nc
import pandas as pd
import numpy as np
```

```python
#cargando dataset.
file = 'data/sst1.cdf'
#file = 'sst2.cdf'
data = nc.Dataset(file)
```

```python
print(data.file_format)
print('\nDIMENSIONES')
print(data.dimensions.keys())
print('\nVARIABLES')
print(data.variables.keys())
print('\nVariable depth')
print(data.variables['T_20'])
```

```python
time_var = data.variables['time']
dtime = nc.num2date(time_var[:],time_var.units)
#ultima fecha: 2018-NOV-10

#Se genera el dframe hasta el 10 NOV.
df = pd.DataFrame(list(data.variables['T_20'][:].data.ravel()), columns = ['SST'], index=dtime)
df
```

### 

### 2. Reemplazando los NaN

```python
# generando el df para añadir.
days_nan = pd.date_range(start='2018-11-11 12:00:00', end='2018-12-31 12:00:00', freq = 'D')
days_nan

nan_df = pd.DataFrame(np.nan, index = days_nan, columns=['SST'])
nan_df

#Calculamos la media.
media = df.SST.mean()
nan_df.SST = nan_df.SST.fillna(media)
nan_df

gp = [df, nan_df]
final = pd.concat(gp)
# DF final de acuerdo a fechas.

final
```

### 3. Serializar / Deserializar


```python
final.to_pickle('data/binary1.pkl')
#Procedemos a importar
datita = pd.read_pickle('data/binary1.pkl')
print('Serializado correctamente!')
datita
```
