```python
#Librerias
import netCDF4 as nc
import pandas as pd
import numpy as np
import xarray as xr
```

```python
#loading dataset.
file = 'data/sst2.cdf'
ds = xr.open_dataset(file)
df = ds.to_dataframe()

missing_na = df['T_20'][13148]
# filling with NaN all missing values.
df.loc[df.T_20==missing_na] = df.loc[df.T_20==missing_na].replace({missing_na : np.nan})

b1 = df.loc[(df.index.get_level_values('lat')==-2) & (df.index.get_level_values('lon')==250.0)]['T_20']
b1 = b1.reset_index(level=0,drop=True)
b1 = b1.reset_index(level=0,drop=True)
b1 = b1.reset_index(level=0,drop=True)


b2 = df.loc[(df.index.get_level_values('lat')==-2) & (df.index.get_level_values('lon')==265.0)]['T_20']
b2 = b2.reset_index(level=0,drop=True)
b2 = b2.reset_index(level=0,drop=True)
b2 = b2.reset_index(level=0,drop=True)


b3 = df.loc[(df.index.get_level_values('lat')==0) & (df.index.get_level_values('lon')==250.0)]['T_20']
b3 = b3.reset_index(level=0,drop=True)
b3 = b3.reset_index(level=0,drop=True)
b3 = b3.reset_index(level=0,drop=True)


b4 = df.loc[(df.index.get_level_values('lat')==0) & (df.index.get_level_values('lon')==265.0)]['T_20']
b4 = b4.reset_index(level=0,drop=True)
b4 = b4.reset_index(level=0,drop=True)
b4 = b4.reset_index(level=0,drop=True)

b5 = df.loc[(df.index.get_level_values('lat')==2) & (df.index.get_level_values('lon')==250.0)]['T_20']
b5 = b5.reset_index(level=0,drop=True)
b5 = b5.reset_index(level=0,drop=True)
b5 = b5.reset_index(level=0,drop=True)


b6 = df.loc[(df.index.get_level_values('lat')==2) & (df.index.get_level_values('lon')==265.0)]['T_20']
b6 = b6.reset_index(level=0,drop=True)
b6 = b6.reset_index(level=0,drop=True)
b6 = b6.reset_index(level=0,drop=True)

df1 = pd.DataFrame({'2S110W': b1, '2S095W': b2 , '0N110W': b3, '0N095W': b4 , '2N110W': b5, '2N095W': b6})
df1
```

```python
new_features = ['2S110W', '2S095W', '0N110W', '0N095W', '2N110W', '2N095W']

# Checking detected null temperatures
for temp_feature in new_features:
    print("Null temperatures in {}: {}".format(temp_feature, df1[temp_feature].isna().sum()))
```

```python
# Adding day and month auxiliary features

period = pd.date_range('20100101 12:00:00','20181231 12:00:00', freq='D')
df1['month'] = period.month
df1['day'] = period.day
df1
```

```python
# Creating an indexing method
def get_data_by_date(dataframe, month, day):
    same_date_data = dataframe[(dataframe.index.month == month) & (dataframe.index.day == day)]
    return same_date_data

df2 = get_data_by_date(df1, 1, 3)
print(df2['2N110W'].sum()/6)
print(df2['2N110W'].mean())
df2
```

```python
# --- MAIN REPLACEMENT PROCESS ---

for temp_feature in new_features: # For each temperature column
    nan_indices = list( df1[temp_feature].isna() ) # Getting data where buoy temp. is nan
    df1.loc[nan_indices,temp_feature] = df1[nan_indices].apply( # And in those places
        lambda d: get_data_by_date(df1, d.month, d.day)[temp_feature].mean(), # Use method to get same day and month temperatures for buoy 'temp_feature', and get the mean
        axis=1) # And do this across rows

df1
```

```python
df2 = df1.drop(['day', 'month'], axis=1)
```
