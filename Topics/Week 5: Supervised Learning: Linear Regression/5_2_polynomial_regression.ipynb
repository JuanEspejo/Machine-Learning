```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('polynomial.csv')

from sklearn.linear_model import LinearRegression
model = LinearRegression()

x = df.x[0:6, np.newaxis] #---convert to 2D array---
y = df.y[0:6, np.newaxis] #---convert to 2D array---

model.fit(x,y)

#---perform prediction---
y_pred = model.predict(x)

#---plot the training points---
plt.scatter(x, y, s=10, color='b')

#---plot the straight line---
plt.plot(x, y_pred, color='r')
plt.show()

#---calculate R-Squared---
print('R-Squared for training set: %.4f' % model.score(x,y))
```

```python
# importing PolynomialFeatures class
from sklearn.preprocessing import PolynomialFeatures

degree = 2
polynomial_features = PolynomialFeatures(degree = degree)

# generating a new feature matrix consisting of all polynomial combinations 
# of the features with a degree of less than or equal to the specified degree
x_poly = polynomial_features.fit_transform(x)

print(polynomial_features.get_feature_names('x'))
print(x_poly)

# using the above generated matrix with the LinearRegression class to train our model
model = LinearRegression()
```

```python
# trining and predicting our model

model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

#---plot the points---
plt.title('degree 2 polynomial model')
plt.scatter(x, y, s=10)

#---plot the regression line---
plt.plot(x, y_poly_pred, color='g')
plt.show()

# printing out the intercept and coefficients of the polynomial function
print('intercept:    ', model.intercept_)
print('coefficients: ', model.coef_)

# printing its R-Squared value
print('\nR-Squared for training set: %.4f' % model.score(x_poly,y))
```
